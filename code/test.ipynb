```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate samples
samples1 = np.random.beta(2, 6, size=100000)
samples2 = np.random.normal(0.25, 0.18, size=100000)

# Step 2: Calculate histograms
hist1, bin_edges1 = np.histogram(samples1, bins=100, density=True)
hist2, bin_edges2 = np.histogram(samples2, bins=100, density=True)

# Step 3: Normalize histograms (already done by density=True)

# Step 4: Calculate overlap
min_bin_edge = max(bin_edges1[0], bin_edges2[0])
max_bin_edge = min(bin_edges1[-1], bin_edges2[-1])
bins = np.linspace(min_bin_edge, max_bin_edge, 100)

hist1_interp = np.interp(bins, (bin_edges1[:-1] + bin_edges1[1:]) / 2, hist1)
hist2_interp = np.interp(bins, (bin_edges2[:-1] + bin_edges2[1:]) / 2, hist2)

overlap = np.minimum(hist1_interp, hist2_interp).sum() * (bins[1] - bins[0])

print(f"Estimated overlap: {overlap:.2%}")

# Optional: Plotting
plt.hist(samples1, bins=100, density=True, alpha=0.5, label='Beta(2, 6)')
plt.hist(samples2, bins=100, density=True, alpha=0.5, label='Beta(3, 4)')
plt.legend()
plt.show()
```
