```python
# normalize names

def remove_accents(input_str):
    # Normalize the input string to decompose accented characters
    normalized_str = unicodedata.normalize('NFD', input_str)
    # Filter out the accent characters
    no_accents_str = ''.join(char for char in normalized_str if unicodedata.category(char) != 'Mn')
    return no_accents_str

def normalize_name(name):
    if name is not None:
        # Remove common suffixes like Jr., Sr., II, III, IV
        suffixes = [" Jr.", " Sr.", " III", " II", " IV", ' I', '.']
        for suffix in suffixes:
            name = name.replace(suffix, "")
        # Remove any extra whitespace
        name = name.strip()
        if name == 'Jeff Wilson' or name == 'Jeffery Wilson':
            name = 'Jeff Wilson'
        if name == 'De\'Von Achane' or name == 'Devon Achane':
            name = 'De\'Von Achane'
        if name == 'Nathaniel Dell' or name == 'Tank Dell':
            name = 'Tank Dell'
        name = remove_accents(name)
    return name

weekly_rosters_2022['player_name'] = weekly_rosters_2022['player_name'].apply(normalize_name)
weekly_rosters_2023['player_name'] = weekly_rosters_2023['player_name'].apply(normalize_name)
adp_2022['Name'] = adp_2022['Name'].apply(normalize_name)
adp_2023['Name'] = adp_2023['Name'].apply(normalize_name)
adp_2024['Name'] = adp_2024['Name'].apply(normalize_name)
weekly_stats_2021['player_display_name'] = weekly_stats_2021['player_display_name'].apply(normalize_name)
weekly_stats_2022['player_display_name'] = weekly_stats_2022['player_display_name'].apply(normalize_name)
weekly_stats_2023['player_display_name'] = weekly_stats_2023['player_display_name'].apply(normalize_name)
k_fantasy_points_df_2023['Name'] = k_fantasy_points_df_2023['Name'].apply(normalize_name)
k_fantasy_points_df_2022['Name'] = k_fantasy_points_df_2022['Name'].apply(normalize_name)
dst_fantasy_points_df_2022['Name'] = dst_fantasy_points_df_2022['Name'].apply(normalize_name)
dst_fantasy_points_df_2023['Name'] = dst_fantasy_points_df_2023['Name'].apply(normalize_name)
weekly_offense_projections_2022['Name'] = weekly_offense_projections_2022['Name'].apply(normalize_name)
weekly_offense_projections_2023['Name'] = weekly_offense_projections_2023['Name'].apply(normalize_name)
weekly_k_projections_2022['Name'] = weekly_k_projections_2022['Name'].apply(normalize_name)
weekly_k_projections_2023['Name'] = weekly_k_projections_2023['Name'].apply(normalize_name)
weekly_dst_projections_2022['Name'] = weekly_dst_projections_2022['Name'].apply(normalize_name)
weekly_dst_projections_2023['Name'] = weekly_dst_projections_2023['Name'].apply(normalize_name)
```

```python
# merge projections 
weekly_offense_projections_2023.rename(columns={'Name': 'player_display_name', 'Week': 'week', 'FantasyPointsPPR': 'projected_fantasy_points_ppr'}, inplace=True)
weekly_k_projections_2023.rename(columns={'Name': 'player_display_name', 'Week': 'week', 'FantasyPointsPPR': 'projected_fantasy_points_ppr'}, inplace=True)
weekly_dst_projections_2023.rename(columns={'Name': 'player_display_name', 'Week': 'week', 'FantasyPointsPPR': 'projected_fantasy_points_ppr'}, inplace=True)

weekly_offense_projections_2022.rename(columns={'Name': 'player_display_name', 'Week': 'week', 'FantasyPointsPPR': 'projected_fantasy_points_ppr'}, inplace=True)
weekly_k_projections_2022.rename(columns={'Name': 'player_display_name', 'Week': 'week', 'FantasyPointsPPR': 'projected_fantasy_points_ppr'}, inplace=True)
weekly_dst_projections_2022.rename(columns={'Name': 'player_display_name', 'Week': 'week', 'FantasyPointsPPR': 'projected_fantasy_points_ppr'}, inplace=True)

projections_2022 = pd.concat([weekly_offense_projections_2022, weekly_k_projections_2022, weekly_dst_projections_2022])
projections_2023 = pd.concat([weekly_offense_projections_2023, weekly_k_projections_2023, weekly_dst_projections_2023])
```
